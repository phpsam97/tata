```python
8. Time Series
pd.to_datetime() – Convert to datetime format.
df['date'].dt.year – Extract the year from a datetime column.
df['date'].dt.month – Extract the month from a datetime column.
df['date'].dt.day – Extract the day from a datetime column.
df['date'].dt.weekday – Extract the weekday.
df['date'].dt.isocalendar() – Get the ISO calendar year, week, and day.
df['date'].dt.month_name() – Get the name of the month.
df['date'].dt.strftime() – Convert a datetime to a string with a custom format.

    
df.resample() – Resample time series data.


df.shift() – Shift data by a specified number of periods.
df.diff() – Calculate the difference between consecutive data points.
```
